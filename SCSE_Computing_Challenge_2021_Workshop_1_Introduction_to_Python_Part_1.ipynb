## **Python Data Types and Variable Assignment**

Variables in programming can store data of different types such as string, integer, etc. These are called data types. In this section we will go through the different datatypes and how they are declared. We will also introduce output formatting for the different datatypes.

```python
# Variable Assignment -- Numeric Data Types

num1 = 2021
num2 = 20.21
num3 = complex(20, 21)
print("num1 =", num1, "\t(Integer)\nnum2 =", num2, 
      "\t(Float)\nnum3 =", num3,"(Complex Number)")
```

More list and tuple methods will be taught in Part 2

Output formatting is done using the `format()` method.

```python
# format() method 
name = 'Jason'
age = 18
height = 170.5
print('Hi! I am {0}. I am {1} years old and {2}cm tall'.format(name, age, height))
```

This is called using positional keyword arguments as the order of objects is followed.

## **Math Operators**


### **Part 1: Arithmetic Operators**
Arithmetic Operators are mathematics functions that take two operands and performs their respective calculations on the operands. 
>- `+`: Addition
- `-`: Subtraction
- `*`: Multiplication
- `/`: Division
- `%`: Modulus
- `**`: Exponent
- `//`: Floor Division

```python
a = 10
b = 20
c = b+a
print(a,"+",b,"=",c)
print(b,"-",a,"=",b-a)
print(a,"*", b,"=",a*b)
print(a,"/", b,"=",a/b) # The division result of a/b is converted to a float value
print(b,"%",a,"=",b%a)
print(a,"**",3,"=",a**3)
print(a,"//",b,"=",a//b) # Converts the division result to an integer
print(b,"//",a,"=",b//a)
```

### **Part 2: Relational Operators**
These operators compares the values of two operands and returns a boolean value corresponding to the result of the comparison.
>- `==` : True if the values of the two operands are equal.
- `!=` : True if the values of the two operands are not equal.
- `<` : True if the value of the right operand is less than the value of the left operand.
- `>`  : True if the value of the right operand is more than the value of the right operand.
- `>=` : True if the value of the right operand is less than or equal to the value of the left operand
- `<=` : True if the value of the left operand is less than or equal to the value of the right operand

```python
a = 10==20
print(a)
```

```python
print(40>30)
```

### **Part 4: Logical Operators**
Logical operators are used on conditional statements(True and False operands).
>- `and` : True if both the operands are true.
- `or`  : True if either of the operands are true.
- `not` : Used to reverse the operator's boolean from True to False or False to True

Nonzero integer values are interpreted as True.

```python
print(True or False)
print(not True)
```

### **Part 5: Member Operators**
This operand checks if the operand on the left of the operand exists in the sequence on the right of the operand.
>- `in`: True if variable found in the specified sequence.
- `not in`: True if variable is not found in the specified sequence.

### **Part 6: Identity Operators**
This operator checks if the operand on the left is the same as the operand on the right.
>- `is` : True if the variables on either side of the operator point to the same object.
- `is not`: True if the variables on either side of the operator do not point to the same object.

## **Conditionals**
Conditional statements in Python perform different set of actions depending on the result of the specified boolean expression.

1. `if` statement:<br>
        if <expression>:
            <statements>
                
2. `if-else` statement:<br>
        if <expression>:
            <statements1>
        else:
            <statements3>
                
3. `if-elif-else` statement:<br>
        if <expression1>:
            <statements1>
        elif <expression2>:
            <statements2>
        else:
            <statements3>               
                
- The expressions need to have a boolean value and hold true for statement to be executed.
- Indentation is used to indicate a block of code, and is very important in python. It indicates whether a block of code belongs to the `if` block, the `elif` block or the `else` block.

## **Loops**
Loops are used to repeat a set of statements 

1. `for` loop:<br>
        for <value> in <list/tuple/dictionary/set/string/>:
            <statements>
The `for` loop is a sentinel controlled loop as we know the number of time the loop will iterate itself before the program is executed.                 
2. `while` loop:<br>
        while <condition>:
               <statements>
In the case of the`while` loop, the loop iterates itself until the loop condition turns false, hence the number of times the loop will be iterated is unknown.<br>
              

- `break`: stops the loop before it has passed through all the items. <br>
                
- `continue` : stops the current iteration of the loop and moves on to the next. <br>
                 
- `pass` : loop remains unaffected

- Indentation is very important as it denotes the set of statements that are going to be iterated over.

PSEUDOCODE:

```python
# range(start, end, step) function - creates a list with integers starting from 'start', 
    # to 'end'-1, adding 'step' to determine the next number in the list.
# The default start value is 0.
# The default step value is 1. 

num = 0
for num in range(10):
    print(num)
```

```python
# for loop 

for num in range(3,10):
    if num==8:
        break;
    elif num==4:
        continue
    elif num == 7:
        pass
    print(num)
```

```python
# while loop

condition = True
num = 0

while(condition):
    if num==10:
        condition = False
    print(num)
    num+=2
```
